## 3. ***Data Wrangling***

### 1.Check for Missing Values

````python
print(data.isnull().sum())  # Display count of missing values
data.dropna(inplace=True)   # Remove missing values (if necessary)
````

### 2.Check for Duplicates

````python
# Duplicate Values Check
print("Duplicate Rows Count:", data.duplicated().sum())
data.drop_duplicates(inplace=True)  # Remove duplicates
````

### 3.Fix Data Types (Convert Columns)

````python
data['release_year'] = data['release_year'].astype(int)  # Convert to integer
data['imdb_score'] = pd.to_numeric(data['imdb_score'], errors='coerce')  # Convert to numeric
````

### 4.Handle Outliers (Optional, for numeric columns)

````python
import numpy as np

Q1 = data['imdb_score'].quantile(0.25)
Q3 = data['imdb_score'].quantile(0.75)
IQR = Q3 - Q1
data = data[(data['imdb_score'] >= (Q1 - 1.5 * IQR)) & (data['imdb_score'] <= (Q3 + 1.5 * IQR))]
````

### 5.Standardize Text Formatting (Genres, Countries, etc.)

````python
data['genres'] = data['genres'].str.lower().str.strip()  # Convert genres to lowercase
data['production_countries'] = data['production_countries'].str.upper()  # Convert to uppercase
````

### 6.Create New Columns (If Required, e.g., Duration in Hours)

````python
data['duration_hours'] = data['runtime'] / 60  # Convert runtime to hours
````

### 7.Export the Cleaned Data

````python
data.to_csv("cleaned_dataset.csv", index=False)  # Save the cleaned dataset
````

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1. IMDb scores for Movies and TV Shows have the same average rating.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (Hâ‚€)**: The mean IMDb scores of Movies and TV Shows are equal.

**Alternative Hypothesis (Hâ‚)**: The mean IMDb scores of Movies and TV Shows are not equal.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import ttest_ind
data_cleaned=pd.read_csv(r"C:\Users\Dell\Downloads\cleaned_dataset.csv")
# Splitting data into two groups
movies = data_cleaned[data_cleaned['type'] == 'Movie']['imdb_score'].dropna()
tv_shows = data_cleaned[data_cleaned['type'] == 'TV Show']['imdb_score'].dropna()

# Performing Independent T-Test
t_stat, p_value = ttest_ind(movies, tv_shows, equal_var=False)
print(f"T-Statistic: {t_stat}, P-Value: {p_value}")
````

##### Which statistical test have you done to obtain P-Value?

We use an independent t-test because we are comparing the mean IMDb scores of two independent groups (Movies vs. TV Shows).

The assumption of equal variance is relaxed using Welchâ€™s t-test (equal_var=False).

##### Why did you choose the specific statistical test?

If p-value < 0.05, we reject Hâ‚€, meaning IMDb scores are significantly different between Movies and TV Shows.

Otherwise, we fail to reject Hâ‚€, meaning no significant difference in ratings

### Hypothetical Statement - 2.  IMDb scores of top 10 production countries are normally distributed.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (Hâ‚€):** IMDb scores for top 10 production countries follow a normal distribution.

**Alternative Hypothesis (Hâ‚):** IMDb scores for top 10 production countries do not follow a normal distribution.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import shapiro

top_10_countries = data_cleaned['production_countries'].value_counts().index[:10]
filtered_data = data_cleaned[data_cleaned['production_countries'].isin(top_10_countries)]

# Shapiro-Wilk Test for normality
shapiro_test = shapiro(filtered_data['imdb_score'].dropna())
print(f"Shapiro-Wilk Test Statistic: {shapiro_test.statistic}, P-Value: {shapiro_test.pvalue}")
````

##### Which statistical test have you done to obtain P-Value?

Shapiro-Wilk test is used because it is a powerful test for normality, especially for small to medium-sized samples.

It helps determine if IMDb scores are normally distributed in the top 10 production countries

##### Why did you choose the specific statistical test?

If p-value < 0.05, we reject Hâ‚€, meaning IMDb scores do not follow a normal distribution.

Otherwise, we fail to reject Hâ‚€, meaning IMDb scores are normally distributed.
